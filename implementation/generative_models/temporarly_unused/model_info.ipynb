```python
from keras.models import load_model
from pathlib import Path
from wgan import WGAN
from IPython.display import SVG
from keras.utils.vis_utils import model_to_dot
from IPython.display import Image
```

```python
folder_name = Path('wgan/2018-04-23_16-29-14')
```

```python
model_dir = folder_name / 'models'
images_dir = folder_name / 'img'
datasets_dir = folder_name / 'generated_datasets'

wgan = WGAN(0, 0, folder_name, images_dir, model_dir, datasets_dir, 0, 0)
model, generator, critic = wgan.load_models()
```

```python
SVG(model_to_dot(model).create(prog='dot', format='svg'))
```

```python
SVG(model_to_dot(generator).create(prog='dot', format='svg'))
```

```python
SVG(model_to_dot(critic).create(prog='dot', format='svg'))
```
